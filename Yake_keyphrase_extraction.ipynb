```python
import json
import yake
```

```python
with open("Chemistrynew.json","r", encoding = "utf-8") as f:
    data = json.load(f)
```

```python
all_content_list = []
for lesson in data['lessons']:
  lesson_name = lesson['lesson_name']
  all_content_list.append(lesson['whole_lesson_content'])
  for sub_topic in lesson['subtopics']:
    sub_topic_name = sub_topic['subtopic_name']
    all_content_list.append(sub_topic['subtopic_content'])
```

```python
text = ' '.join(all_content_list)
```

```python
kw_extractor = yake.KeywordExtractor(n=3)  # Set n to the desired value (e.g., 3 for unigrams, bigrams, and trigrams)
keywords = kw_extractor.extract_keywords(text)
```

```python
for kw in keywords:
    print(kw)
```
